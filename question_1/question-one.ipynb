```python
mean_income = data['Income-wages'].mean()
median_income = data['Income-wages'].median()

print(f"Mean Yearly Income: ${mean_income:,.2f}")
print(f"Median Yearly Income: ${median_income:,.2f}")
```

b)

```python
std_dev_income = data['Income-wages'].std()
q1 = data['Income-wages'].quantile(0.25)
q3 = data['Income-wages'].quantile(0.75)
iqr = q3 - q1

print(f"Standard Deviation of Yearly Income: ${std_dev_income:,.2f}")
print(f"Interquartile Range (IQR): ${iqr:,.2f}")
```

```python
# If there is no variability in yearly income, all individuals would have the same income. In this case:

# The standard deviation would be 0 (no spread).
# The IQR would also be 0 (since the data would be constant).
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Boxplot for Yearly Income by Education Level
plt.figure(figsize=(10, 6))
sns.boxplot(x='Education', y='Income-wages', data=data)
plt.title('Yearly Income by Education Level')
plt.xticks(rotation=45)
plt.show()

# Boxplot for Yearly Income by Region
plt.figure(figsize=(10, 6))
sns.boxplot(x='Region_recode', y='Income-wages', data=data)
plt.title('Yearly Income by Region')
plt.xticks(rotation=45)
plt.show()

```

```python
#Explanation of Yearly income by education level boxplot.
      #People with higher levels of education have higher levels of yearly income, this means they earn more.
      #People with lower levels of education have higher levels of yearly income, this means they earn less income.
      #In conclusion, the higher the education level, the higher the income level, the lower the education level,
#the lower the income level.
```

```python
#Explanation of Yearly income by region boxplot.
      #The south  region has lowest level of yearly income.
      #The midwest region has lower level of yearly income.
      #The west region has higher level of yearly income.
      #The Northeast region has the highest level of yearly income.
```

```python
# Group by Education Level and calculate summary statistics
education_summary = data.groupby('Education')['Income-wages'].describe()

# Group by Region and calculate summary statistics
region_summary = data.groupby('Region_recode')['Income-wages'].describe()

print(education_summary)
print(region_summary)
```
