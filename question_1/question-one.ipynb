```python
mean_income = data['Income-wages'].mean()
median_income = data['Income-wages'].median()

print(f"Mean Yearly Income: ${mean_income:,.2f}")
print(f"Median Yearly Income: ${median_income:,.2f}")
```

b)

```python
std_dev_income = data['Income-wages'].std()
q1 = data['Income-wages'].quantile(0.25)
q3 = data['Income-wages'].quantile(0.75)
iqr = q3 - q1

print(f"Standard Deviation of Yearly Income: ${std_dev_income:,.2f}")
print(f"Interquartile Range (IQR): ${iqr:,.2f}")
```
